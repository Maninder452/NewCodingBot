```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
data = pd.read_csv('study_performance.csv')
```

```python
# Select two features for scatter plot
feature1 = 'math_score'
feature2 = 'reading_score'
```

```python
# Plot the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=feature1, y=feature2, data=data, hue='test_preparation_course', palette='viridis')
plt.title('Scatter Plot of {} vs {}'.format(feature1, feature2))
plt.xlabel(feature1)
plt.ylabel(feature2)
plt.legend(title='Test Preparation Course', loc='upper left')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Select the features for clustering
X = data[['math_score', 'reading_score']]
```

```python
# 1. Elbow Method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks(np.arange(1, 11, 1))
plt.show()
```

```python
# 2. Silhouette Score
silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    silhouette_scores.append(silhouette_score(X, kmeans.labels_))

plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Score')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.xticks(np.arange(2, 11, 1))
plt.show()
```

```python
from sklearn.metrics import davies_bouldin_score
```

```python
# Initialize an empty list to store Davies-Bouldin Index scores
db_scores = []
```

```python
# Calculate Davies-Bouldin Index for different values of K
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    db_scores.append(davies_bouldin_score(X, kmeans.labels_))
```

```python
# Plot the Davies-Bouldin Index scores
plt.figure(figsize=(10, 6))
plt.plot(range(2, 11), db_scores, marker='o')
plt.title('Davies-Bouldin Index')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Davies-Bouldin Index')
plt.xticks(np.arange(2, 11, 1))
plt.show()
```
